# Exploratory Sales Analysis

This notebook demonstrates exploratory data analysis on the sample sales dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../backend/sample_data/sample_sales_data.csv')
df.head()
```

## Basic Statistics

```python
df.describe()
```

## Revenue by Category

```python
df['line_total'] = df['quantity'] * df['unit_price'] - df['discount_amount']
category_revenue = df.groupby('product_category')['line_total'].sum().sort_values(ascending=False)
category_revenue.plot(kind='bar', title='Revenue by Category')
plt.ylabel('Revenue ($)')
plt.show()
```

## Sales by Country

```python
country_revenue = df.groupby('customer_country')['line_total'].sum().sort_values(ascending=False)
country_revenue.plot(kind='barh', title='Revenue by Country', color='steelblue')
plt.xlabel('Revenue ($)')
plt.show()
```

## Monthly Trend

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['month'] = df['order_date'].dt.to_period('M')
monthly = df.groupby('month')['line_total'].sum()
monthly.plot(title='Monthly Revenue Trend', marker='o')
plt.ylabel('Revenue ($)')
plt.show()
```
